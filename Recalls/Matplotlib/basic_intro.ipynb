```python
import matplotlib.pyplot as plt #importing matplotlib for plotting
import numpy as np #importing numpy for numerical operations
import pandas as pd #importing pandas for data manipulation
```

```python
Key terms to know in Matplotlib:

* Data Point : A single observation or measurement in a dataset, represented by coordinates (x, y) in a 2D plot.
* X-Axis/Y-Axis : The horizontal (X) and vertical (Y) reference lines in a plot that define the coordinate system.
* Axes : The area where data is plotted, including ticks, labels, and grid lines.
* Figure : The entire window or page that contains the Axes and other elements like titles and legends.
* Plot : A single graph or chart created using Matplotlib.
* Marker : A symbol used to represent data points in a plot, such as circles, squares, or triangles.
* Line-Style : The appearance of lines in a plot, including solid, dashed, or dotted lines.
* Color : The hue used to differentiate data series or elements in a plot.
* Legend : A key that explains the symbols, colors, or line styles used in a plot.
* Title : A heading that describes the content of a plot. 
* Grid : A network of lines that helps to visually align data points in a plot.
* Function : A reusable block of code that performs a specific task, such as plotting data.
* Method : A function that is associated with an object, such as a plotting method for a Figure or Axes object.
* Parameter : An input value that customizes the behavior of a function or method.
* Keyword Argument : A parameter specified by name when calling a function, allowing for more readable code.
* Object-Oriented API : A programming approach that uses objects and methods to create plots, providing more control and flexibility.
* DPI (Dots Per Inch) : A measure of the resolution of a plot, indicating how many dots fit into one inch.
* Backend : The rendering engine used by Matplotlib to create plots, which can vary based on the output format (e.g., PNG, PDF).
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x, y) # Simple line plot with x as x-axis and y as y-axis
plt.show() # Display the plot
```

```python
x = ["Aaditya", "Bobby", "Chitra", "Divya", "Esha"]
y = [23, 45, 12, 67, 34]
plt.plot(x, y, marker='o', linestyle='--', color='r', label='Age Data')
plt.title('Age of Students')
plt.xlabel('Names')
plt.ylabel('Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
* plt.plot() is a function in the Matplotlib library used to create 2D line plots. It takes in data points for the x-axis and y-axis and generates a visual representation of the relationship between them. The function can be customized with various parameters such as marker style, line style, color, labels, and titles to enhance the appearance and clarity of the plot. 
      marker : Specifies the style of the data points in the plot (e.g., 'o' for circles).
      linestyle : Defines the style of the line connecting the data points (e.g., '--' for dashed lines).
      color : Sets the color of the line and markers (e.g., 'r' for red).
      label : Provides a label for the data series, which can be displayed in a legend.
* plt.show() is a function in the Matplotlib library that displays the current figure or plot. 
*plt.title() adds a title to the plot.
* plt.xlabel() and plt.ylabel() set the labels for the x-axis and y-axis, respectively.
* plt.legend() displays a legend on the plot, which helps identify different data series based on their labels.
* plt.grid() adds a grid to the plot, improving readability by aligning data points with horizontal and vertical lines.
```
